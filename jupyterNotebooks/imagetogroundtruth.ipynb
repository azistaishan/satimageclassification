```python
def decimal_coords(coords, ref):
    decimal_degrees = coords[0] + coords[1] / 60 + coords[2] / 3600
    if ref == "S" or ref =='W' :
        decimal_degrees = -decimal_degrees
    return decimal_degrees

def getLatitudeLongitude(fileFpath):
    fileName= Path(fileFpath).name
    # print(fileName)
    with open(fileFpath, 'rb') as src:
        img = Image(src)
    try:
        lat = decimal_coords(img.gps_latitude, img.gps_latitude_ref)
        long = decimal_coords(img.gps_longitude, img.gps_longitude_ref)
        return ({
            'name': fileName,
            'lat': lat,
            'lat_raw': img.gps_latitude,
            'lat_ref': img.gps_latitude_ref,
            'long': long,
            'long_raw': img.gps_longitude,
            'long_ref': img.gps_longitude_ref,
            'time_taken': img.datetime,
            'img_path': fileFpath,
        }) 
    except:
        return({
            'name': fileName,
            'time_taken': img.datetime,
            'img_path': fileFpath
        })
def dictocsv(fileList):
    outdict = {} 
    for i in fileList:
        out = getLatitudeLongitude(i) 
        outkeys = out.keys()
        # print(f"1 here {outkeys}")
        # if len(keys<outkeys):
        #     print(f"Keys to be extended: {keys}")
        #     keys.extend(outkeys)
        #     print(f"Keys extended: {keys}")
        # i = 0
        if 'lat' not in outkeys:
            pass
            #print("Lat is not in outkeys")
            #print(len(outkeys))
        else:
            for key in outkeys:
                # print(f"2 here {key}")
                if key in outdict.keys():
                    # print(f'{key} is in dictionary')
                    outdict[key].append(out[key])
                else:
                    outdict[key] = [out[key]]
                    #outdict[key] = out[key]
                # print(outdict)
    for key in outdict:
        print(f"values in key: {key:10} = {len(outdict[key])}")    
    return outdict
    
```
